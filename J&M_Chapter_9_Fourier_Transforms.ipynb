```python
import numpy as np
import matplotlib.pyplot as plt
```

Screenshot 2023-12-05 at 11.37.25 AM.png

```python
N = 10
pi = np.pi
e = np.e
j = 1j

n = (pi/2 * np.linspace(0,N,100))

ePower = lambda k: -j*2*(pi/N)*k*n
```

```python
x = np.sin(n)
plt.plot(x)
```

```python
X = lambda k: np.sum(x * np.power(e, ePower(k)))
```

```python
X(1)
```

```python
plt.polar([X(k) for k in n])
```

```python
plt.polar(x * np.power(e, ePower(n[80])))
```
